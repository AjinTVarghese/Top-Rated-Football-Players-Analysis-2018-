```python
#import libraries
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv(r'fifa_players.csv')
```

```python
df.head(2)
```

```python
df.shape
```

```python
#it contains more than 17954 rows we have to clean that so lets do that
```

```python
df2 = df.drop(['body_type','penalties','short_passing','curve','long_passing','freekick_accuracy','heading_accuracy','crossing','national_jersey_number','long_shots','reactions','sprint_speed','national_team_position','national_rating','release_clause_euro','international_reputation(1-5)','weak_foot(1-5)','age','full_name','balance','agility', 'height_cm', 'weight_kgs', 'composure','jumping','standing_tackle','aggression','positioning','marking','vision','wage_euro','potential','value_euro'], axis=1, errors='ignore')
```

```python
df2.head(2)
```

```python
df2['birth_date'] = pd.to_datetime(df2['birth_date'])
```

```python
df2['year'] = df2['birth_date'].apply(lambda x: x.strftime('%Y'))
```

```python
df2 = df2.drop(['birth_date','volleys','shot_power'], axis=1, errors='ignore')
```

```python
df2.head(2)
```

```python
#age of the Player
```

```python
df2['year']=df2['year'].astype(int)
```

```python
df2['year']  = 2024-df2['year']
```

```python
df2.head(2)
```

```python
df2=df2[df2['year']<40]
```

```python
df2.shape
```

```python
#cleanup null values
```

```python
df2.isna().sum()
```

```python
df2[df2['national_team'].isna()==1]
```

```python
#removed players who are not in the national_team
```

```python
df2=df2.dropna()
```

```python
df2.shape
```

```python
#we dont want gk in this letms remove it
```

```python
df2 = df2[df2['positions'] != 'GK']
```

```python
df2.head(2)
```

```python
df2.shape
```

```python
#we want only best players rating over 80
```

```python
df2=df2[df2['overall_rating']>80]
```

```python
df2 = df2.drop(['nationality','skill_moves(1-5)','sliding_tackle'], axis=1, errors='ignore')
```

```python
df2.shape
```

```python
df2.head()
```

```python
df2=df2.rename(columns={'name':'Name',
                        'positions':'Played Positions',
                        'overall_rating':'Rating',
                        'preferred_foot':'Strong Foot',
                        'national_team':'National Team',
                        'finishing':'Finishing',
                        'dribbling':'Dribbling',
                        'ball_control':'Ball Control',
                        'acceleration':'Acceleration',
                        'stamina':'Stamina',
                        'strength':'Strength',
                        'interceptions':'Interceptions',
                        'year':'Age'
                        })
```

```python
df2.head(2)
```

```python
#reorder columns
```

```python
df2=df2[['Name','Age','Rating','Played Positions','National Team','Strong Foot','Finishing','Dribbling','Ball Control','Acceleration','Stamina','Strength','Interceptions']]
```

```python
df2.head(2)
```

```python
df2= df2.sort_values(by='Rating', ascending=False)
```
